**2. Count how many commits were made in the past week.**

```python
import os
from dotenv import load_dotenv
import json
import requests
import re
load_dotenv()
from datetime import datetime, timedelta
```

```python
now = datetime.now().isoformat()
last_week = datetime.now() - timedelta(days=7)
print(now)
print(last_week)
```

```python
token = os.getenv('token')
```

**1. Obtain all the commits made in the past week via API, which is a JSON array that contains multiple commit objects.**

```python
url = "https://api.github.com/repos/ironhack-datalabs/datamad1020-rev/commits"
headers = {"Authorization": f"token {token}"}
parameters = {'since': last_week  , 'until': now}
res = requests.get(url, headers=headers, params = parameters)
```

```python
type(res)
```

```python
res.status_code
```

```python
count_json = res.json()
```
