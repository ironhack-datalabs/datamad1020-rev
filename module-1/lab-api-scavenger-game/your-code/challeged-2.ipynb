```python
from datetime import datetime,timedelta
def get_n_days_ago_str(N):
    N = 7
    date_N_days_ago = datetime.now()- timedelta(days = N)
    since_str= date_N_days_ago.strftime("%Y-%m-%d")
    return since_str
```

```python
since_str = get_n_days_ago_str(7)
```

```python
parameters = {'since':since_str}
```

```python
url_c = f'https://api.github.com/repos/{owner}/{repo}/commits'
response_c = requests.get(url = url_c,
                          params =parameters )
```
