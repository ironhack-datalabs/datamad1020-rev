```python
# Challenge 1: Fork Languages
```

```python
import os
from dotenv import load_dotenv
import json
import requests
import re
load_dotenv()
```

```python
token = os.getenv("token")
```

```python
headers = {"Authorization": f"token {token}"}
```

```python
url = "https://api.github.com/repos/ironhack-datalabs/datamad1020-rev/"
```

```python
forks = url + "forks"
```

```python
response = requests.get(forks, headers=headers)
```

```python
response.status_code
```

```python
data = response.json()
```

```python
print(data)
```

```python
type(data)
```

```python
for i in data:
    print(i["language"])
```

```python
# Challenge 2: Count Commits
```

```python
url = "https://api.github.com/repos/ironhack-datalabs/datamad1020-rev/"
```

```python
commit = url + "commits"
```

```python
response2 = requests.get(commit, headers=headers)
```

```python
response2.status_code
```

```python
data2 = response2.json()
```

```python
print(data2)
```

```python
for i in data2:
    if str(i) > "2020-10-26" < "2020-11-01":
        print(i["commit"])
```
