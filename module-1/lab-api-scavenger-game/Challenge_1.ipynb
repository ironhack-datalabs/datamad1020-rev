***1. Obtain the full list of forks created from the main lab repo via Github API.***

```python
url = "https://api.github.com/repos/ironhack-datalabs/datamad1020-rev/forks"
headers = {"Authorization": f"token {token}"}
res = requests.get(url, headers=headers)
```

```python
eljeison = res.json()
```

```python
forks = []
for item in eljeison:
    forks.append(item["full_name"])

print(forks)
print("There is a total of", len(forks), "forks made of the" ,eljeison[0]["name"], "repository")
```

**2. Loop the JSON response to find out the language attribute of each fork. Use an array to store the language attributes of each fork.**

```python
print(eljeison)
```

```python
language_attributes = []
for item in eljeison:
    if item["language"] not in language_attributes:
        language_attributes.append((item["full_name"],item["language"]))

print(language_attributes)
```

**3: Print the language array. It should be something like:**

```python
languages = []
for item in eljeison:
    #print(item["languages_url"])
    url = item["languages_url"]
    res = requests.get(url, headers=headers)
    lang_json = res.json()
    for lang in lang_json.keys():
        if lang not in languages:
            languages.append(lang)
            
print(languages)
```
