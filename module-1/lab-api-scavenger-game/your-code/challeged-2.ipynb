```python
import os
from dotenv import load_dotenv
import json
import requests
import re
load_dotenv() 
```

```python
token = os.getenv('token')
```

```python
header = {'Authorization' : f'token {token}'}
```

```python
repo = 'datamad1020-rev'
owner ='ironhack-datalabs'
```

```python
url_c = f'https://api.github.com/repos/{owner}/{repo}/commits'
response_c = requests.get(url_c,headers = header)
```

```python
result_c = response_c.json()
result_c 
```

```python
past_week_commit = []

for elem in result_c:
    date =  elem['commit']['committer']['date'][:10]
    if '2020-10-26'< date <'2020-11-01':
        past_week_commit.append(elem)
past_week_commit
commit_total = len(past_week_commit)    
commit_total
    
```
